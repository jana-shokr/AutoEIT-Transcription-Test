```python
import os
import whisper
import pandas as pd

# 1. Setup paths
data_dir = "data"
audio_folder = os.path.join(data_dir, "AutoEIT Test Files", "Sample Audio Files and Transcriptions")

# Get all the audio files
audio_files = [f for f in os.listdir(audio_folder) if f.endswith(('.wav', '.mp3', '.m4a'))]

# 2. Make sure the model is loaded 
print("Loading 'small' model...")
model = whisper.load_model("small")

# 3. Create an empty list to hold our final results
results_list = []

# 4. The Loop! Process every file one by one
print(f"Starting batch transcription of {len(audio_files)} files...")

for file in audio_files:
    print(f"-> Transcribing {file}...")
    file_path = os.path.join(audio_folder, file)
    
    
    result = model.transcribe(file_path)
    
    # Save the filename and the text together
    results_list.append({
        "Filename": file,
        "Transcription": result["text"].strip()
    })

# 5. Turn the results into a clean Pandas frame
print("\n--- DONE! Here is the final data ---")
df = pd.DataFrame(results_list)

# Display the table in Jupyter
display(df)
```
